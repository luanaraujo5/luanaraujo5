```python
import osmnx as ox
import networkx as nx 
import matplotlib.pyplot as plt

# Carregamos o mapa de Manaus
G = ox.graph_from_address('Manaus, Amazonas, Brasil', dist=8000, network_type='drive')

# Definimos a origem (lat, long)
origin = (-3.13256, -60.01824)

# Definimos o destino (lat, long)
destination = (-3.08309, -60.02831)

# Encontramos o node mais próximo da (lat, long) informado, para a origem e destino
origin_node = ox.get_nearest_node(G, origin)
destination_node = ox.get_nearest_node(G, destination)

# Identificamos a rota mais curta utilizando dijkstra
route = nx.shortest_path(G, origin_node, destination_node, method='dijkstra', weight='length')

# Apresentamos o mapa com a rota
fig, ax = ox.plot_graph_route(G, route, route_linewidth=6, node_size=0, bgcolor='k')


```

```python
distancia = nx.shortest_path_length(G, origin_node, destination_node, method='dijkstra', weight='length')
print('A distância entre os pontos é de {:0.2f} metros'.format(distancia))
```
